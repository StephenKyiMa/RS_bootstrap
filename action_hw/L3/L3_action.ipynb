```python
# from mlxtend.frequent_patterns import apriori
# from mlxtend.frequent_patterns import association_rules
from efficient_apriori import apriori
df = pd.read_csv('/Users/mafangqi/My_Study/rs 后厂/Code/L3/MarketBasket/Market_Basket_Optimisation.csv', header=None)
# df.info()
```

```python
df.fillna(0,inplace=True)
df.head()
```

```python
transactions = []
for i in range(0,len(df)):
    tran_set = set()
    for j in range(0,len(df.columns)):
        if str(df.values[i,j])!='0':
            tran_set.add(str(df.values[i,j]))
    transactions.append(tran_set)

transactions
```

```python
itemsets, rules = apriori(transactions, min_support=0.02, min_confidence=0.2)
print('Frequency Set is', itemsets)
print('-'*130)
print('Related Rule is', rules)
```

```python
f = str(itemsets)
def word_cloud(f):
    cut_text = word_tokenize(f)
    cut_text = " ".join(cut_text)
    wc = WordCloud(max_words=100, width=2000, height=1200)
    wordcloud = wc.generate(cut_text)
    wordcloud.to_file('/Users/mafangqi/My_Study/rs 后厂/Code/L3/MarketBasket/wordcloud.jpg')
word_cloud(f)
```
